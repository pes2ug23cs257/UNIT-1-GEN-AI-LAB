```python
from transformers import pipeline
```

```python
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)
```

```python
movie_plot = """
A group of friends plan a vacation in a remote forest cabin.
Strange events begin to occur at night as a mysterious presence
starts hunting them one by one.
"""

```

```python
candidate_genres = [
    "Action",
    "Comedy",
    "Romance",
    "Horror",
    "Science Fiction"
]
```

```python
result = classifier(
    movie_plot,
    candidate_labels=candidate_genres
)

result

```

```python
for label, score in zip(result["labels"], result["scores"]):
    print(f"{label}: {score:.2f}")
```
